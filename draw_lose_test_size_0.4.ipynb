```python
import pandas as pd
```

```python
data1 = pd.read_csv('soccer.csv', sep = ',', header = 0, engine = 'python')
data2 = pd.read_csv('soccer2.csv', sep = ',', header = 0, engine = 'python')
data3 = pd.read_csv('soccer3.csv', sep = ',', header = 0, engine = 'python')
```

```python
# csv 파일 병합
total_data = pd.concat([data1, data2, data3])
total_data
```

```python
total_data.columns = total_data.columns.str.replace(' ', '_')
total_data.columns
```

```python
# 전반전 슈팅값 변수 추가
total_data['Shots_1st_Half'] = total_data['Shots_Off_1st_Half'] + total_data['Shots_On_1st_Half']
```

```python
# homeAway에서 home은 1, Away는 0으로 설정
total_data = total_data.replace({'homeAway': 2}, 0)
total_data
```

```python
# 인덱스 재설정
total_data = total_data.reset_index(drop=True)
total_data
```

```python
# 독립변수와 종속변수 추출
total_data = total_data[['homeAway', 'Ball_Possession_1st_Half', 'Corners_1st_Half', 'Fouls_Commited_1st_Half', 'Goals_1st_Half', 'Red_Cards_1st_Half', 'Shots_1st_Half', 'Shots_On_1st_Half', 'Yellow_Cards_1st_Half', 'Result']]
total_data
```

```python
#무, 패 데이터 추출
idx = total_data[total_data['Result'] == "W"].index
total_data.drop(idx, inplace = True)
total_data
```

```python
# 무승부는 1로, 패배는 0으로 값 변경
total_data = total_data.replace({'Result': 'D'}, 1)
total_data = total_data.replace({'Result': 'L'}, 0)
total_data
```

```python
#결측치 확인
total_data.info()
```

```python
#데이터 내보내기
total_data.to_csv('soccer_data.csv', index = False)
```

```python
#타겟 데이터 따로 저장
target = total_data['Result']
#total_data에서 종속 변수 제거
total_data = total_data.drop(columns=['Result'], axis = 1)
```

```python
total_data.head()
```

```python
#표준화 작업
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
soccer_half_scaled = scaler.fit_transform(total_data)
```

```python
print(soccer_half_scaled[0])
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
# X, Y 설정
Y = target
X = soccer_half_scaled
```

```python
# 훈련 데이터와 테스트 데이터 분리
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.4, random_state = 0, stratify = Y)
```

```python
print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

```python
#target 자료형 확인
Y_train.info()
```

```python
Y_test.info()
```

```python
print(round(sum(Y_train)/len(Y_train)*100, 2), '%')
```

```python
print(round(sum(Y_test)/len(Y_test)*100, 2), '%')
```

```python
lr_soccer_half = LogisticRegression(max_iter=10000) #max_iter 확인
lr_soccer_half.fit(X_train, Y_train)
```

```python
Y_predict = lr_soccer_half.predict(X_test)
```

```python
lr_soccer_half.coef_
```

```python
lr_soccer_half.intercept_
```

로지스틱 회귀모형 평가

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
```

```python
# 오차행렬 확인 (예측치와 실제값 비교)
confusion_matrix(Y_test, Y_predict)
```

```python
accuracy = accuracy_score(Y_test, Y_predict)
precision = precision_score(Y_test, Y_predict)
recall = recall_score(Y_test, Y_predict)
f1 = f1_score(Y_test, Y_predict)
```

```python
print('정확도: {0:.3f}, 정밀도: {1:.3f}, 재현율: {2:.3f}, F1: {3:.3f}'.format(accuracy, precision, recall, f1))
```
