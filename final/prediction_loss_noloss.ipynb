```python
import pandas as pd
```

```python
# 36-38라운드에 대한 패/무패 예측
testdata = pd.read_csv('3638round.csv', sep = ',', header = 0, engine = 'python')
```

```python
testdata
```

```python
# 승(W)과 무(D)를 무패(NL)로 변경. sf 칼럼 삭제
testdata = testdata.replace({'result': 'W'}, 'NL')
testdata = testdata.replace({'predict': 'W'}, 'NL')
testdata = testdata.replace({'result': 'D'}, 'NL')
testdata = testdata.replace({'predict': 'D'}, 'NL')
testdata = testdata.drop(columns = ['sf'], axis = 1)
```

```python
df = pd.DataFrame(testdata)
df
```

```python
# 패, 무패 예측 성공률
print(round(len(df.loc[df['predict'] == df['result']])/len(df)*100, 2), '%')
```
