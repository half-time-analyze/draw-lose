```python
import pandas as pd
```

```python
data1 = pd.read_csv('soccer.csv', sep = ',', header = 0, engine = 'python')
data2 = pd.read_csv('soccer2.csv', sep = ',', header = 0, engine = 'python')
data3 = pd.read_csv('soccer3.csv', sep = ',', header = 0, engine = 'python')
```

```python
# csv 파일 병합
total_data = pd.concat([data1, data2, data3])
total_data
```

```python
total_data.columns = total_data.columns.str.replace(' ', '_')
total_data.columns
```

```python
# 전반전 슈팅값 변수 추가
total_data['Shots_1st_Half'] = total_data['Shots_Off_1st_Half'] + total_data['Shots_On_1st_Half']
```

```python
# homeAway에서 home은 1, Away는 0으로 설정
total_data = total_data.replace({'homeAway': 2}, 0)
total_data
```

```python
# 독립변수와 종속변수 추출
total_data = total_data[['homeAway', 'Ball_Possession_1st_Half', 'Corners_1st_Half', 'Fouls_Commited_1st_Half', 'Goals_1st_Half', 'Red_Cards_1st_Half', 'Shots_1st_Half', 'Shots_On_1st_Half', 'Yellow_Cards_1st_Half', 'Result']]
total_data
```

```python
#무, 패 데이터 추출
idx = total_data[total_data['Result'] == "W"].index
total_data.drop(idx, inplace = True)
total_data
```

```python
# 무승부는 1로, 패배는 0으로 값 변경
total_data = total_data.replace({'Result': 'D'}, 1)
total_data = total_data.replace({'Result': 'L'}, 0)
total_data
```
